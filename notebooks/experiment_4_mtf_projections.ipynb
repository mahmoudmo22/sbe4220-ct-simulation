# Experiment 4: Spatial Resolution vs Angular Sampling

Quantifying how angular undersampling degrades MTF.

```python
import os, sys
sys.path.append(os.path.abspath("../src"))
import numpy as np
import matplotlib.pyplot as plt
from phantom import generate_qa_phantom
from forward_projection import generate_sinogram
from fbp import reconstruct_fbp
from mtf import compute_mtf_from_reconstruction
from utils import plot_mtf_curves, display_images_grid
```

## 1. Reconstruct at different projection counts

```python
phantom, metadata = generate_qa_phantom(size=256)
angles_list = [360, 180, 90, 45]
recons = {}
mtf_results = {}

for num_angles in angles_list:
    sino, angles = generate_sinogram(phantom, num_angles)
    recon = reconstruct_fbp(sino, angles, filter_name="ram-lak")
    recons[num_angles] = recon
    freq, mtf_vals = compute_mtf_from_reconstruction(recon, method="edge")
    mtf_results[f"{num_angles} Angles"] = (freq, mtf_vals)
```

## 2. Plot Results

```python
display_images_grid([recons[a] for a in angles_list], titles=[f"{a} Angles" for a in angles_list], ncols=2, save_path="../results/exp4_visual.png")
plot_mtf_curves(mtf_results, title="MTF vs Angular Sampling", save_path="../results/exp4_mtf.png")
```
